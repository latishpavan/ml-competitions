```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import os
from sklearn.linear_model import LogisticRegression
```

```python
pwd = os.getcwd()
files = os.listdir(pwd)
csv_files = filter(lambda x: '.csv' in x, files)

from functools import partial
read_csv = partial(pd.read_csv, index_col=False, encoding="ISO-8859-1")
df_dict = {csv : read_csv(csv) for csv in csv_files if csv.startswith('t')}
```

```python
train_df = df_dict['train.csv']
test_df  = df_dict['test.csv']
```

```python
null_values = train_df.isnull().sum()
```

```python
test_df['Survived'] = 1
test_df.isTrain = False
train_df.isTrain = True

total_data = pd.concat([train_df, test_df], axis=0)
```

```python
filter_data = total_data.drop(remove_cols, axis=1)
```
